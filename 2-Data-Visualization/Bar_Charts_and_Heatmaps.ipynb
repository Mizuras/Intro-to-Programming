```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

## **Load the data**
Memuat dataset menggunakan perintah pd.read_csv.

```python
# Path of the file to read
fastfood_filepath = "/content/fastfood_calories.csv"
# Read the file into a variable fastfood_data
fastfood_data = pd.read_csv(fastfood_filepath, index_col="restaurant")
```

fastfood_filepath = "/content/fastfood_calories.csv": Baris ini mendefinisikan path atau jalur (path) ke file CSV yang berisi data yang ingin dibaca. Path ini disimpan dalam variabel fastfood_filepath. Dalam hal ini, path menuju ke file "fastfood_calories.csv" di direktori "/content/".

fastfood_data = pd.read_csv(fastfood_filepath, index_col="restaurant"): Baris ini membaca file CSV yang diberikan (fastfood_filepath) ke dalam sebuah DataFrame menggunakan fungsi pd.read_csv() dari pustaka Pandas. Parameter index_col="restaurant" digunakan untuk menentukan kolom "restaurant" sebagai indeks baris dari DataFrame yang dibuat. DataFrame ini kemudian disimpan dalam variabel fastfood_data.

## **Bar chart**
Katakanlah kita ingin membuat diagram batang yang menunjukkan penundaan kedatangan rata-rata untuk penerbangan Spirit Airlines (kode maskapai: NK), berdasarkan bulan.

```python
# Atur lebar dan tinggi gambar
plt.figure(figsize=(10,6))

# Tambahkan judul
plt.title("Tingkat caloires dalam makanan, by Restaurant")

# Bar chart menunjukkan setiap calories dari tiap menu
sns.barplot(x=fastfood_data.index, y=fastfood_data['total_carb'], color='Red')

# Add label for vertical axis
plt.ylabel("Jumlah calories (dalam angka)")
```

1. plt.figure(figsize=(10,6)): Baris ini mengatur ukuran
gambar yang akan digunakan untuk menampilkan bar chart. Ukuran yang ditentukan adalah 10 inci lebarnya dan 6 inci tingginya.

2. plt.title("Tingkat caloires dalam makanan, by Restaurant"): Ini menambahkan judul ke bar chart.

3. sns.barplot(x=fastfood_data.index, y=fastfood_data['total_carb'], color='Red'): Ini adalah baris yang sebenarnya membuat bar chart menggunakan Seaborn. Ini menggunakan data dari DataFrame fastfood_data. x=fastfood_data.index menentukan sumbu x sebagai indeks DataFrame, dan y=fastfood_data['total_carb'] menentukan sumbu y sebagai kolom "total_carb". Parameter color='Red' mengatur warna bar menjadi merah.

4. plt.ylabel("Jumlah calories (dalam angka)"): Ini menambahkan label pada sumbu vertikal (y-axis).

## **Heatmap**
Kami memiliki satu jenis plot lagi untuk dipelajari: peta panas!

Di sel kode di bawah, kita membuat peta panas untuk memvisualisasikan pola di fastfood_data dengan cepat. Setiap sel diberi kode warna sesuai dengan nilainya.

```python
# Atur lebar dan tinggi gambar
plt.figure(figsize=(14,7))

# Tambahkan judul
plt.title("Tingkat komposisi dalam makanan, by Restaurant")

# Peta panas menunjukkan tingkat calories untuk setiap restaurant
sns.heatmap(data=fastfood_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Komposisi")
```
