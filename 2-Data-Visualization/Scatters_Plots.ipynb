# Color-coded scatter plots
Kita dapat menggunakan plot sebar untuk menampilkan hubungan antara (bukan dua, tapi...) tiga variabel! Salah satu cara untuk melakukan ini adalah dengan memberi kode warna pada titik-titik tersebut.

Misalnya, untuk memahami bagaimana merokok mempengaruhi hubungan antara BMI dan biaya asuransi, kita dapat memberi kode warna pada titik-titik tersebut dengan 'perokok', dan memplot dua kolom lainnya ('bmi', 'biaya') pada sumbunya.

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

Plot sebar ini menunjukkan bahwa walaupun orang yang bukan perokok cenderung membayar sedikit lebih mahal dengan meningkatnya BMI, orang yang merokok membayar JAUH lebih banyak.

Untuk lebih menekankan fakta ini, kita dapat menggunakan perintah sns.lmplot untuk menambahkan dua garis regresi, yang sesuai dengan perokok dan bukan perokok. (Anda akan melihat bahwa garis regresi untuk perokok memiliki kemiringan yang jauh lebih curam dibandingkan dengan garis regresi untuk bukan perokok!)

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

Perintah sns.lmplot di atas bekerja sedikit berbeda dari perintah yang telah Anda pelajari sejauh ini:

- Daripada mengatur x=insurance_data['bmi'] untuk memilih kolom 'bmi' di Insurance_data, kita mengatur x="bmi" untuk menentukan nama kolom saja.
- Demikian pula, y="charges" dan hue="smoker" juga berisi nama kolom.
- Kami menentukan kumpulan data dengan data=insurance_data.

Terakhir, ada satu plot lagi yang akan Anda pelajari, yang mungkin terlihat sedikit berbeda dari biasanya Anda melihat plot sebar. Biasanya, kami menggunakan plot sebar untuk menyorot hubungan antara dua variabel kontinu (seperti "bmi" dan "biaya"). Namun, kita dapat mengadaptasi desain plot sebar untuk menampilkan variabel kategori (seperti "perokok") pada salah satu sumbu utama. Kami akan merujuk pada tipe plot ini sebagai plot sebar kategorikal, dan kami membangunnya dengan perintah sns.swarmplot.

```python
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```

Plot ini antara lain menunjukkan kepada kita bahwa:

- rata-rata, orang yang bukan perokok dikenakan tarif lebih rendah dibandingkan perokok, dan
- pelanggan yang membayar paling banyak adalah perokok; sedangkan pelanggan yang membayar paling sedikit adalah bukan perokok.
